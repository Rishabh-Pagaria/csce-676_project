## Dataset download (Socrata paging, reliable for large datasets)

Why:
- Avoids timeouts / partial downloads on large open-data tables
- Keeps memory stable and makes the pipeline reproducible

````python
DATASET_ID = "6vp6-wxuq"  # CDC 500 Cities 2019 release
SOCRATA_CSV_ENDPOINT = f"https://data.cdc.gov/resource/{DATASET_ID}.csv"

def fetch_socrata_csv(endpoint: str, page_size: int = 50000, max_rows: int | None = None) -> pd.DataFrame:
    """
    Fetch Socrata CSV endpoint with paging via $limit/$offset.

    Args:
      endpoint: Socrata resource endpoint ending in .csv
      page_size: rows per request (50k is a good balance)
      max_rows: optional cap for quick iteration; set None for full pull

    Returns:
      DataFrame with all fetched rows.
    """
    chunks = []
    offset = 0
    total = 0

    while True:
        limit = page_size
        if max_rows is not None:
            remaining = max_rows - total
            if remaining <= 0:
                break
            limit = min(limit, remaining)

        params = {"$limit": limit, "$offset": offset}
        r = requests.get(endpoint, params=params, timeout=60)
        r.raise_for_status()

        # Read this page into a DataFrame
        chunk = pd.read_csv(pd.io.common.StringIO(r.text))
        if chunk.empty:
            break

        chunks.append(chunk)
        got = len(chunk)
        total += got
        offset += got

        print(f"Fetched {got:,} rows (total {total:,})")

        # Stop if last page smaller than page_size (end of data)
        if got < limit:
            break

    df = pd.concat(chunks, ignore_index=True) if chunks else pd.DataFrame()
    return df

# For checkpoint A/B/C/D, it's usually fine to download full data.
# If you want fast iteration first, set max_rows=200000 then remove the cap.
df_raw = fetch_socrata_csv(SOCRATA_CSV_ENDPOINT, page_size=50000, max_rows=None)

print("Shape:", df_raw.shape)
df_raw.head()

# Test 1: Dataset must not be empty and should contain substantial rows
assert df_raw.shape[0] > 100000, "Dataset appears too small — download may have failed."
assert df_raw.shape[1] >= 20, "Unexpected schema — missing expected columns."

print("Test 1 passed: Dataset size and structure validated.")

# Test 2: Ensure required columns exist and key column contains valid data
required_columns = [
    "year", "geographiclevel", "measure",
    "data_value", "tractfips"
]

for col in required_columns:
    assert col in df_raw.columns, f"Missing required column: {col}"

# Check that Census Tract level exists in dataset
assert "Census Tract" in df_raw["geographiclevel"].unique(), \
    "Census Tract level not found — unexpected dataset contents."

print("Test 2 passed: Required columns and geographic level verified.")

````


## Lock schema + safe cleaning prep

Why: Now that we know the exact columns, we hardcode them to avoid fragile inference and prevent downstream failures.


````python
# =========================
# Column constants + cleaning
# =========================
COL_YEAR = "year"
COL_STATE = "stateabbr"
COL_CITY  = "cityname"
COL_GEOLEVEL = "geographiclevel"
COL_MEASURE = "measure"
COL_MEASURE_ID = "measureid"
COL_VALUE = "data_value"
COL_VALUE_TYPE = "data_value_type"
COL_VALUE_TYPE_ID = "datavaluetypeid"
COL_TRACT = "tractfips"
COL_CITYFIPS = "cityfips"
COL_GEOLOC = "geolocation"
COL_POP = "populationcount"
COL_CATEGORY = "category"
COL_SHORT_Q = "short_question_text"

# Fail fast if schema changes (prevents silent downstream errors)
required = [
    COL_YEAR, COL_STATE, COL_CITY, COL_GEOLEVEL, COL_MEASURE, COL_VALUE,
    COL_VALUE_TYPE, COL_VALUE_TYPE_ID, COL_TRACT
]
missing = [c for c in required if c not in df_raw.columns]
if missing:
    raise ValueError(f"Dataset schema mismatch. Missing columns: {missing}")

df = df_raw.copy()

# IDs should be identifiers (integers), not floats.
df[COL_TRACT] = pd.to_numeric(df[COL_TRACT], errors="coerce").round().astype("Int64")
df[COL_CITYFIPS] = pd.to_numeric(df[COL_CITYFIPS], errors="coerce").round().astype("Int64")

# Normalize text columns to avoid duplicate categories due to whitespace
for c in [COL_CITY, COL_MEASURE, COL_VALUE_TYPE, COL_CATEGORY, COL_STATE, COL_GEOLEVEL]:
    df[c] = df[c].astype(str).str.strip()

print("Cleaned dtypes (key columns):")
display(df[[COL_TRACT, COL_CITYFIPS, COL_YEAR, COL_VALUE]].dtypes)

print("\nMissingness after tract/cityfips conversion:")
display(df[[COL_TRACT, COL_CITYFIPS, COL_VALUE]].isna().mean().mul(100).round(2))

display(df[COL_GEOLEVEL].value_counts().head(20))
display(df[COL_VALUE_TYPE].value_counts().head(20))
display(df[COL_YEAR].value_counts().sort_index())

# Test 1: tractfips should remain mostly present and have substantial uniqueness
tract_missing_pct = df[COL_TRACT].isna().mean() * 100
tract_unique = df[COL_TRACT].nunique(dropna=True)

assert tract_missing_pct < 10, f"Too many missing tract IDs after conversion: {tract_missing_pct:.2f}%"
assert tract_unique > 10000, f"Unexpectedly low number of unique tracts: {tract_unique}"

print("Test 1 passed: tractfips conversion looks valid.")

# Test 2: expected analysis categories exist (prevents later filter failures)
geolevels = set(df[COL_GEOLEVEL].unique())
value_types = set(df[COL_VALUE_TYPE].unique())
years = set(df[COL_YEAR].unique())

assert "Census Tract" in geolevels, "Expected geographiclevel 'Census Tract' not found."
assert "Crude prevalence" in value_types, "Expected data_value_type 'Crude prevalence' not found."
assert 2017 in years, "Expected year 2017 not found."

print("Test 2 passed: key categorical values for slicing are present.")
````

# Pivot to Tract x Measure Matrix

Why:
- Clustering, PCA, and anomaly detection require one row per tract and one column per measure.

````python
wide = df_slice.pivot_table(
    index=COL_TRACT,
    columns=COL_MEASURE,
    values=COL_VALUE,
    aggfunc="mean"  # Safe aggregation if duplicates exist
)

print("Wide shape (tracts x measures):", wide.shape)

# Sanity check
print("Expected tracts:", df_slice[COL_TRACT].nunique())
print("Actual rows in wide:", wide.shape[0])

# Check missingness per measure
missing_per_measure = (wide.isna().mean() * 100).sort_values(ascending=False).round(2)
display(missing_per_measure)

wide.head()

# Test 1: Wide matrix should not unexpectedly drop a large fraction of tracts
expected_tracts = df_slice[COL_TRACT].nunique(dropna=True)
actual_tracts = wide.shape[0]

drop_rate = (expected_tracts - actual_tracts) / expected_tracts
assert drop_rate < 0.05, f"Too many tracts dropped during pivot: {drop_rate:.2%} (likely missing tractfips)."

print("Test 1 passed: pivot retains most tracts.")

# Test 2: Expect stable measure coverage and low missingness in the wide matrix
n_measures = wide.shape[1]
total_missing = int(wide.isna().sum().sum())
missing_pct = total_missing / (wide.shape[0] * wide.shape[1]) * 100 if wide.size else 100

assert 15 <= n_measures <= 30, f"Unexpected # of measures in wide matrix: {n_measures} (expected ~20)."
assert missing_pct < 5, f"Wide matrix too sparse: {missing_pct:.2f}% missing."

print("Test 2 passed: feature coverage and sparsity validated.")
````

# PCA Exploration

Why:
- Determine intrinsic dimensionality
- Identify how many latent health factors explain most variance
- Decide whether clustering should be done in reduced space



````python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Fit PCA on scaled data
pca = PCA()
pca.fit(X_scaled)

explained = pca.explained_variance_ratio_
cum_explained = np.cumsum(explained)

# Plot cumulative variance
plt.figure(figsize=(6,4))
plt.plot(range(1, len(cum_explained)+1), cum_explained, marker='o')
plt.axhline(y=0.80, linestyle='--')
plt.axhline(y=0.90, linestyle='--')
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("PCA Cumulative Explained Variance")
plt.grid(True)
plt.show()

# Create summary table
pca_table = pd.DataFrame({
    "Component": range(1, len(explained)+1),
    "Explained Variance Ratio": explained,
    "Cumulative Variance": cum_explained
})

display(pca_table.head(10))

# How many components to reach 90%?
n_90 = np.argmax(cum_explained >= 0.90) + 1
print(f"Number of components needed for 90% variance: {n_90}")

# Test 1: Explained variance ratios must be within [0, 1] and sum ~ 1
assert np.all((explained >= 0) & (explained <= 1)), "Invalid explained variance ratios (outside [0,1])."

ev_sum = explained.sum()
assert abs(ev_sum - 1.0) < 1e-8, f"Explained variance ratios do not sum to 1 (sum={ev_sum})."

print("Test 1 passed: explained variance ratios are valid and sum to ~1.")

# Test 2: Cumulative explained variance must be non-decreasing and end near 1
assert np.all(np.diff(cum_explained) >= -1e-12), "Cumulative explained variance is not monotonic."

cum_end = float(cum_explained[-1])
assert abs(cum_end - 1.0) < 1e-8, f"Cumulative variance does not end near 1 (end={cum_end})."

print("Test 2 passed: cumulative variance is monotonic and ends near 1.")

````

This dataset does NOT behave like 20 independent health measures.

Instead:

- There are a few strong latent health factors.
- Many measures move together.
- There is heavy multicollinearity.

This strongly justifies:

- Dimensionality reduction before clustering.
- That clustering in full 20D space may introduce redundancy.
